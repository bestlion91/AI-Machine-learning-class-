```python
import pandas as pd
```

```python
df=pd.read_csv('IMDB_Movies_Dataset.csv') # Bu kod kompyuteringizdagi IMDB_Movies_Dataset.csv faylini o‘qib, uni df nomli DataFrame (jadval) sifatida yuklaydi.
# pd — bu pandas kutubxonasi (import pandas as pd bo'lishi kerak oldin).
# read_csv() — CSV formatidagi faylni o‘qish uchun funksiyadir.
# 'IMDB_Movies_Dataset.csv' — fayl nomi
```

```python
df=df.dropna(subset='Writer', axis=0) # Bu kod Writer ustunida bo‘sh (NaN) qiymat bor bo‘lgan qatorlarni (row) o‘chiradi.
# dropna() - Bu metod NaN (bo‘sh) qiymatlarni o‘chirish uchun ishlatiladi.
# subset='Writer' - Faqat Writer ustunini tekshiradi. Agar faqat shu ustunda NaN bo‘lsa — o‘chiradi.
# axis=0 - Bu — qatorlar (rows) ni anglatadi. (standart qiymat, yozmasangiz ham bo‘ladi)
# df = ... - Natijani asl df ga qayta yozib saqlab qo‘yadi.
```

```python
df.info()
```

```python
df.isnull().sum() #Bu kod sizning df (DataFrame) jadvalingizda qaysi ustunda nechta NaN (bo‘sh) qiymat borligini ko‘rsatadi.
# df.isnull() - Jadvaldagi har bir katak uchun True yoki False qaytaradi:
                                                                        #True → agar qiymat NaN bo‘lsa
                                                                        #False → agar qiymat to‘liq bo‘lsa
# .sum()- Har bir ustundagi True larni sanaydi (ya'ni, bo‘sh qiymatlar sonini).
```

```python
df['Metascore']=df['Metascore'].fillna(df['Metascore'].mean()) #Bu kod Metascore ustunidagi bo‘sh (NaN) qiymatlarni o‘sha ustundagi o‘rtacha qiymat (mean) bilan to‘ldiradi.
# df['Metascore'] - DataFrame ichidagi 'Metascore' ustunini oladi.
# .mean() - NaN bo‘lmagan qiymatlarning aritmetik o‘rtachasini hisoblaydi. Masalan: [70, 80, NaN, 90] → o‘rtacha = (70 + 80 + 90) / 3 = 80
# .fillna(...) - NaN joylarni shu o‘rtacha qiymat bilan to‘ldiradi.
# df['Metascore'] = ... - Yangi qiymatlar bilan ustunni yangilaydi.
```

```python
df.head(50)
```

```python
metascore_mean = df['Metascore'].mean() #Bu kod Metascore ustunidagi mavjud (ya'ni NaN bo'lmagan) qiymatlarning aritmetik o‘rtachasini hisoblab chiqaradi va uni chop etadi.
print("Metascore ustunining o'rtacha qiymati:", metascore_mean)
```

```python
df['Budget'].fillna(df['Budget'].mode()[0], inplace=True) #'mode yordamida to'ldirish. Budget ustunidagi NaN (bo‘sh) qiymatlarni eng ko‘p uchraydigan qiymat bilan to‘ldiradi.
# df['Budget']- Budget ustunini tanlayapti.
# .mode() - Bu Budget ustunidagi **eng ko‘p uchraydigan qiymat(lar)**ni qaytaradi.
# [0] - Ba'zida .mode() bir nechta qiymat qaytarishi mumkin (agar ular bir xil darajada ko‘p uchrasa). [0] orqali birinchisi tanlanadi.
# .fillna(..., inplace=True) - NaN qiymatlarni tanlangan qiymat bilan to‘ldiradi va inplace=True esa o‘zgarishni asosiy DataFrameda saqlaydi (ya’ni alohida tayinlash kerak emas).
```

```python
budget_mode=df['Budget'].mode() # **eng ko‘p uchraydigan qiymat(lar)**ni topadi (bu moda, o‘rtacha emas!).
print("Budget ustunining o'rtacha qiymati:", budget_mode)
```
