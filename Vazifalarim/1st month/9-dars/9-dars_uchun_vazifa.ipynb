```python
df=df.dropna(subset=['Type of crop'],axis=0)
```

```python
for col in df.columns:#(Har bir ustunni navbatma-navbat olayapti)
   if df[col].isnull().sum()>0:# (Agar ustunda bo‘sh qiymatlar (NaN) mavjud bo‘lsa:)
        if df[col].dtype=='object':#(Ustun turi object (ya'ni, matn) bo‘lsa:)
            df[col].fillna(df[col].mode()[0],inplace=True)# (Eng ko‘p uchraydigan qiymat (mode()[0]) bilan to‘ldir) 
        else:
            df[col].fillna(df[col].mean(),inplace=True)#(Aks holda (ya'ni ustun sonli bo‘lsa), o‘rtacha qiymat bilan to‘ldir)
```
