# Frequence Encoding

```python
df['Ticket'].value_counts()
```

```python
frequency_encoding=df['Ticket'].value_counts().to_dict() #Ticket ustunidagi har bir qiymat nechta marta uchraganini hisoblaydi.
# to_dict() - Natijani lug‘at (dictionary) shaklga o‘giradi:
df['Ticket_encoded']=df['Ticket'].map(frequency_encoding) # map() har bir Ticket qiymatini yuqoridagi lug‘atdan topib, o‘sha raqam bilan almashtiradi.
#Bu kod Ticket ustunidagi qiymatlarni nechta marta takrorlanganiga qarab raqamlaydi (ya'ni, frekvensiya bo‘yicha kodlaydi) va yangi ustun Ticket_encodedga saqlaydi
```

```python
df.info()
```

```python
df.drop(('Ticket'),axis=1,inplace=True)

# df.drop(...) — DataFrame dan ustun yoki qatorni o‘chiradi.
# 'Ticket' — o‘chiriladigan ustun nomi.
# axis=1 — bu ustun (column) ekanligini bildiradi. (axis=0 bo‘lsa, qator bo‘ladi).
# inplace=True — o‘zgarishni to‘g‘ridan-to‘g‘ri df ichida saqlaydi, yangi nusxa yaratmaydi.

# Frequency Encoding — bu kategoriyali ustundagi qiymatlarni ularning qanchalik tez-tez uchrashiga qarab raqam bilan almashtirish usuli.
# Ya'ni, har bir kategoriya qiymatiga:
# u datasetda necha marta uchragan bo‘lsa, o‘sha son biriktiriladi.
```
